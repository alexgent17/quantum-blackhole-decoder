```python
# ==============================================================================
# 🌌 PYRAMIDAL LIGHTNING PROTOCOL: THE ULTIMATE QUANTUM BREAKOUT
# Framework: PennyLane (High-Performance Lightning Engine)
# Strategy: Multi-Scale Long-Range Entanglement & Pyramidal Ansatz (180 Params)
# ==============================================================================

import pennylane as qml
from pennylane import numpy as qnp
import matplotlib.pyplot as plt
import time

n_qubits = 10
dev = qml.device("lightning.qubit", wires=n_qubits)

# Pure message initialization (Target)
message_state = qnp.array([1.0 / qnp.sqrt(2), 1.0 / qnp.sqrt(2)])

print("⚡ Pyramidal Quantum Core Engaged. Hardware Acceleration: MAX.")
print("🌌 Target: Force the 1024-dimensional space past the 70% boundary.\n")

# ==============================================================================
# 1. SCRAMBLER BACKBONE (THE BLACK HOLE)
# ==============================================================================
def advanced_scrambler(message_state):
    qml.StatePrep(message_state, wires=0)
    for i in range(n_qubits - 1):
        qml.IsingXX(1.2, wires=[i, i+1])
        qml.IsingYY(1.0, wires=[i, (i+2) % n_qubits])

# ==============================================================================
# 2. PYRAMIDAL LONG-RANGE ANSATZ (6 LAYERS - 180 PARAMETERS)
# ==============================================================================
@qml.qnode(dev, diff_method="parameter-shift")
def cost_function(weights):
    advanced_scrambler(message_state)

    idx = 0
    # Scaled to 6 Deep Layers for multi-scale renormalization
    for layer in range(6):

        # A. LONG-RANGE INTERACTION HACK (10 params per layer)
        # Connects opposite sides of the 10-qubit ring to compress the Hilbert space
        for i in range(n_qubits):
            jump = (i + 3 + layer) % n_qubits  # Dynamic jumping structure per layer
            qml.IsingXY(weights[idx], wires=[i, jump])
            idx += 1

        # B. DUAL-AXIS ROTATION BLOCK (20 params per layer)
        for qubit in range(n_qubits):
            qml.RX(weights[idx], wires=qubit)
            qml.RY(weights[idx+1], wires=qubit)
            idx += 2

    return qml.expval(qml.Projector([1], wires=9))

# ==============================================================================
# 3. HIGH-PRECISION ADAPTIVE OPTIMIZATION
# ==============================================================================
print("🌀 Launching Pyramidal Adam Optimization...")

# 6 layers * (10 long-range + 20 rotation params) = 180 parameters
num_params = 180
weights = qnp.random.uniform(low=-0.005, high=0.005, size=(num_params,), requires_grad=True)

# Refined stepsize for deep parameter tuning
optimizer = qml.AdamOptimizer(stepsize=0.06, beta1=0.95, beta2=0.99)
steps = 45
loss_history = []

start_time = time.time()

for step in range(steps):
    step_start = time.time()

    weights, cost = optimizer.step_and_cost(cost_function, weights)
    loss_history.append(cost)

    step_end = time.time()
    current_fidelity = 1.0 - cost

    print(f"Step {step+1:2d}/{steps} | Fidelity: {current_fidelity:.2%} | Time: {step_end - step_start:.1f}s")

end_time = time.time()
print(f"\n📊 TOTAL EXECUTION TIME: {(end_time - start_time)/60:.2f} minutes.")

# Plot the definitive breakout trajectory
plt.figure(figsize=(8, 4.5))
plt.plot(loss_history, color='#d90429', linewidth=2.5, label='Pyramidal 180-Param Path')
plt.title('🚀 The Ultimate Quantum Breakout (10 Qubits)', fontsize=12, fontweight='bold')
plt.xlabel('Steps', fontsize=10)
plt.ylabel('Cost Value', fontsize=10)
plt.grid(True, alpha=0.2)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
